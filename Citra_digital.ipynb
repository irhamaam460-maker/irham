```python

# upload file dari komputer / HP
from google.colab import files
uploaded = files.upload()
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('nailong.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Negative
negative = 255 - img

# Logaritmik
img_float = img.astype(np.float32)
c = 255 / np.log(1 + np.max(img_float))
log_image = c * np.log(1 + img_float)
log_image = np.array(log_image, dtype=np.uint8)

# Power Law (Gamma)
gamma = 0.5
power = np.array(255*(img/255)**gamma, dtype='uint8')

# Display
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title("Gambar Asli")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(negative)
plt.title("Negative")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(log_image)
plt.title("Logaritmik")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(power)
plt.title("Power Law (Gamma)")
plt.axis("off")

plt.show()
```
